Retrieve and save MetaNetX annotations for metabolites used in the MetaFlux SBML exports.

```python
# use -python-local-only-non-strict when calling PathwayTools
import pythoncyc
# from pythoncyc.PToolsFrame import Symbol, PFrame
# from pythoncyc.PTools import sendQueryToPTools, PToolsError
# from IPython.display import display, HTML
import cobra
import json
import urllib
import time
cobra.Configuration.solver = 'glpk'
pythoncyc.all_orgids()
```

```python
model = cobra.io.read_sbml_model("lumbricus_rubellus/MetaFlux/SBML_export.xml")
all_mets = set()
for m in model.metabolites:
    met_id = m.annotation.get('biocyc')
    if met_id:
        all_mets.add(met_id[met_id.index(':')+1:])
    else:
        print(m.id)
len(all_mets)
```

POLYMER_INST contains instances of species classes involved e.g. in beta-oxidation

```python
model = cobra.io.read_sbml_model("lumbricus_terrestris/MetaFlux/SBML_export.xml")
for m in model.metabolites:
    met_id = m.annotation.get('biocyc')
    if met_id:
        all_mets.add(met_id[met_id.index(':')+1:])
    else:
        print(m.id)
len(all_mets)
```

```python
metacyc = pythoncyc.select_organism('meta')
for m in all_mets:
    try:
        print(m, metacyc.get_slot_values(m, "DBLINKS").get("|BIGG|"), metacyc.get_slot_values(m, "DBLINKS").get("|METANETX|"))
    except:
        print(m, "?")
```

MetaNetX unification links in MetaCyc can be outdated!

```python
metanetx_annotations = dict()
```

```python
query = r"https://rdf.metanetx.org/sparql/?default-graph-uri=https%3A%2F%2Frdf.metanetx.org%2F&query=PREFIX+mnx%3A+%3Chttps%3A%2F%2Frdf.metanetx.org%2Fschema%2F%3E%0D%0APREFIX+owl%3A+%3Chttp%3A%2F%2Fwww.w3.org%2F2002%2F07%2Fowl%23%3E%0D%0APREFIX+rdf%3A+%3Chttp%3A%2F%2Fwww.w3.org%2F1999%2F02%2F22-rdf-syntax-ns%23%3E%0D%0APREFIX+rdfs%3A+%3Chttp%3A%2F%2Fwww.w3.org%2F2000%2F01%2Frdf-schema%23%3E%0D%0APREFIX+metacycC%3A+%3Chttps%3A%2F%2Fidentifiers.org%2Fmetacyc.compound%3A%3E%0D%0ASELECT+%3Fxref%0D%0AWHERE+%7B%0D%0A++++%3Fmetabolite+a+mnx%3ACHEM+.%0D%0A++++%3Fmetabolite+mnx%3AchemXref+metacycC%3ACCCCCC+.%0D%0A++++%3Fmetabolite+mnx%3AchemXref+%3Fxref%0D%0A%7D%0D%0A&format=application%2Fsparql-results%2Bjson"
for m in all_mets:
    if m in metanetx_annotations and len(metanetx_annotations[m]): # allow retries
        continue
    try:
        result = urllib.request.urlopen(query.replace('CCCCCC', m.strip()))
        data = result.read()
        result = json.loads(data.decode(result.info().get_content_charset('utf-8')))
        metanetx_annotations[m] = [r['xref']['value'] for r in result['results']['bindings']]
        time.sleep(0.1)
    except:
        metanetx_annotations[m] = []
        print(m)
```

```python
[k for k,v in metanetx_annotations.items() if len(v) == 0]
```

```python
with open("metanetx_annotations.json", "w") as fp:
    json.dump(metanetx_annotations, fp)
```
